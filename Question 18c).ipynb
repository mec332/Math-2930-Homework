```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u(x, t):
    return (np.exp(-((np.pi**2)/40)*t)) * ((160/(np.pi**2))* np.sin(np.pi/2) *np.sin((np.pi/40)*x))


x = np.linspace(-6, 6, 30)
t = np.linspace(-6, 6, 30)

X, T = np.meshgrid(x, t)
Z = u(X, T)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, T, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('T');
```
